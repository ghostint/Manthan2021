```python
def filter_data(model , tweets):
    tw_c = 1
    tweets = []
    for sd in some_data:
    #     print(f"Tweet: {sd}\n")
        bigs, trigs, third = model.predict(sd)
        yep = [x for x in third if x[1] > 0.05]
        if len(yep) >= 1:
            tweets.append((sd, third))
            tw_c += 1
    return tw_c, tweets

    
    
```

```python
def filter_by_sentiment(tweets):
    pos = []
    neg = []
    
    for tweet in tweets:
        senti = sia.polarity_scores(tweet)
        if senti["neg"] > 0.3:
            neg.append(tweet)
        if senti["pos"] > 0.3:
            pos.append(tweet)
            
    return [pos,neg]
    
    
        
```
